```python
import pandas as pd
import numpy as np
import time as t
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
```

```python
dataframe_project = pd.read_csv(r'D:\College Education\Semester 5\19CSE305 Machine Learning\Project\Coding\patches_gabor_15816_1.csv')
dataframe_project
```

```python
from sklearn import preprocessing as prepro
label_encoder = prepro.LabelEncoder()
dataframe_project['class'] = label_encoder.fit_transform(dataframe_project['class'])
dataframe_project
```

```python
input_project = dataframe_project.loc[:, dataframe_project.columns[1:25]]
output_project = dataframe_project.loc[:, dataframe_project.columns[25]]

input_train, input_test, output_train, output_test = train_test_split(input_project, output_project, test_size=0.3, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
input_train_res, output_train_res = sm.fit_resample(input_train, output_train.ravel())
```

```python
#randomForest
# Project data classification rules
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from itertools import product
rf_model = RandomForestClassifier()
param_grid={
    'n_estimators': [10, 20, 30, 50, 100]
}
scores_rf = cross_val_score(rf_model, input_project.values, output_project.values, cv=10)
accuracy_list_rf = []
precision_list_rf = []
recall_list_rf = []
f1_list_rf = []

start_train_rf = t.time()
best_accuracy_rf = 0
best_params_rf = {}
for params in product(*param_grid.values()):
    param_dict = dict(zip(param_grid.keys(), params))
    rf_loop = RandomForestClassifier(**param_dict)
    rf_loop.fit(input_train_res, output_train_res)
    predicted_value = rf_loop.predict(input_train_res)

    score = accuracy_score(output_train_res, predicted_value)
    precision = precision_score(output_train_res, predicted_value, average='weighted')
    recall = recall_score(output_train_res, predicted_value, average='weighted')
    f1 = f1_score(output_train_res, predicted_value, average='weighted')
    
    accuracy_list_rf.append(score)
    precision_list_rf.append(precision)
    recall_list_rf.append(recall)
    f1_list_rf.append(f1)
    
    print(f"Parameters: {param_dict}, Accuracy: {score}")

    if score > best_accuracy_rf:
        best_accuracy_rf = score
        best_params_rf = param_dict

rf_best_model = RandomForestClassifier(**best_params_rf)
rf_best_model.fit(input_train_res, output_train_res)
predicted_rf_train = rf_best_model.predict(input_train_res)
end_train_rf = t.time()
training_time_rf = end_train_rf - start_train_rf

start_test_rf = t.time()
predicted_rf_test = rf_best_model.predict(input_test)
end_test_rf = t.time()
testing_time_rf = end_test_rf - start_test_rf

mean_accuracy_rf = sum(accuracy_list_rf) / len(accuracy_list_rf)
mean_precision_rf = sum(precision_list_rf) / len(precision_list_rf)
mean_recall_rf = sum(recall_list_rf) / len(recall_list_rf)
mean_f1_rf = sum(f1_list_rf) / len(f1_list_rf)
mean_training_time_rf = training_time_rf / len(scores_rf)
```

```python
print("Training data")
print(classification_report(output_train_res, predicted_rf_train))
print("Accuracy: ", accuracy_score(output_train_res, predicted_rf_train))
print("Precision: ", precision_score(output_train_res, predicted_rf_train, average='weighted'))
print("Recall: ", recall_score(output_train_res, predicted_rf_train, average='weighted'))
print("F1-score: ", f1_score(output_train_res, predicted_rf_train, average='weighted'))
print(training_time_rf, 's')
print("Testing data")
print(classification_report(output_test, predicted_rf_test))
print("Accuracy: ", accuracy_score(output_test, predicted_rf_test))
print("Precision: ", precision_score(output_test, predicted_rf_test, average='weighted'))
print("Recall: ", recall_score(output_test, predicted_rf_test, average='weighted'))
print("F1-score: ", f1_score(output_test, predicted_rf_test, average='weighted'))
print(testing_time_rf, 's')
print(best_params_rf)
print('Mean accuracy: ', mean_accuracy_rf)
print('Standard deviation: ',scores_rf.std())
print('Mean Precsion: ', mean_precision_rf)
print('Mean Recall: ', mean_recall_rf)
print('Mean F1-score: ', mean_f1_rf)
```
