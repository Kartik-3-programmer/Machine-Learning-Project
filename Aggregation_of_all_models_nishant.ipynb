```python
import pandas as pd
import numpy as np
import time as t
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

```python
dataframe_project = pd.read_csv(r'D:\College Education\Semester 5\19CSE305 Machine Learning\Project\Coding\patches_gabor_15816_1.csv')
dataframe_project
```

```python
dataframe_project.isna().sum()
```

```python
from sklearn import preprocessing as prep
label_encoder = prep.LabelEncoder()
dataframe_project['class'] = label_encoder.fit_transform(dataframe_project['class'])
dataframe_project
```

```python
input_project = dataframe_project.loc[:, dataframe_project.columns[1:25]]
output_project = dataframe_project.loc[:, dataframe_project.columns[25]]

input_train, input_test, output_train, output_test = train_test_split(input_project, output_project, test_size=0.3, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
input_train_res, output_train_res = sm.fit_resample(input_train, output_train.ravel())
```

```python
# Decision Tree classification
from sklearn.tree import DecisionTreeClassifier
from itertools import product
dt_model = DecisionTreeClassifier()
param_grid={
    'max_depth': [1, 2, 5, 10, 15],
    'max_features': ['sqrt', 'log2'],
    'max_leaf_nodes': [2, 5, 10, 15, 20],
    'criterion': ['gini', 'entropy', 'log_loss']
}
scores_dt = cross_val_score(dt_model, input_project.values, output_project.values, cv=10)
accuracy_list_dt = []
precision_list_dt = []
recall_list_dt = []
f1_list_dt = []

start_train_dt = t.time()
best_accuracy_dt = 0
best_params_dt = {}
for params in product(*param_grid.values()):
    param_dict = dict(zip(param_grid.keys(), params))
    dt_loop = DecisionTreeClassifier(**param_dict)
    dt_loop.fit(input_train_res, output_train_res)
    predicted_value = dt_loop.predict(input_train_res)

    score = accuracy_score(output_train_res, predicted_value)
    precision = precision_score(output_train_res, predicted_value, average='weighted')
    recall = recall_score(output_train_res, predicted_value, average='weighted')
    f1 = f1_score(output_train_res, predicted_value, average='weighted')
    
    accuracy_list_dt.append(score)
    precision_list_dt.append(precision)
    recall_list_dt.append(recall)
    f1_list_dt.append(f1)
    
    print(f"Parameters: {param_dict}, Accuracy: {score}")

    if score > best_accuracy_dt:
        best_accuracy_dt = score
        best_params_dt = param_dict

dt_best_model = DecisionTreeClassifier(**best_params_dt)
dt_best_model.fit(input_train_res, output_train_res)
predicted_dt_train = dt_best_model.predict(input_train_res)
end_train_dt = t.time()
training_time_dt = end_train_dt - start_train_dt

start_test_dt = t.time()
predicted_dt_test = dt_best_model.predict(input_test)
end_test_dt = t.time()
testing_time_dt = end_test_dt - start_test_dt

mean_accuracy_dt = sum(accuracy_list_dt) / len(accuracy_list_dt)
mean_precision_dt = sum(precision_list_dt) / len(precision_list_dt)
mean_recall_dt = sum(recall_list_dt) / len(recall_list_dt)
mean_f1_dt = sum(f1_list_dt) / len(f1_list_dt)
```

```python
print("Training data")
print(classification_report(output_train_res, predicted_dt_train))
print("Accuracy: ", accuracy_score(output_train_res, predicted_dt_train))
print("Precision: ", precision_score(output_train_res, predicted_dt_train, average='weighted'))
print("Recall: ", recall_score(output_train_res, predicted_dt_train, average='weighted'))
print("F1-score: ", f1_score(output_train_res, predicted_dt_train, average='weighted'))
print(training_time_dt, 's')
print("Testing data")
print(classification_report(output_test, predicted_dt_test))
print("Accuracy: ", accuracy_score(output_test, predicted_dt_test))
print("Precision: ", precision_score(output_test, predicted_dt_test, average='weighted'))
print("Recall: ", recall_score(output_test, predicted_dt_test, average='weighted'))
print("F1-score: ", f1_score(output_test, predicted_dt_test, average='weighted'))
print(testing_time_dt, 's')
print(best_params_dt)
print("Mean Accuracy: ", mean_accuracy_dt)
print("Standard Deviation: ",scores_dt.std())
print('Mean Precsion: ', mean_precision_dt)
print('Mean Recall: ', mean_recall_dt)
print('Mean F1-score: ', mean_f1_dt)
```
